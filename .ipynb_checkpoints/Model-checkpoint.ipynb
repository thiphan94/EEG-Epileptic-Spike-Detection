```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
p = np.fromfile('exp1-train-400pos.bin', dtype=np.float32).reshape([400, 5, 768])
print(p[0])
```

```python
# cas positive
y = p[0][0]
y2 = p[0][1]
y3 = p[0][2]
y4 = p[0][3]
y5 = p[0][4]
# Plot lists and show them
plt.figure(1, figsize=(16, 4)) 
plt.plot(y, 'red')
plt.plot(y2, 'blue')
plt.plot(y3, 'violet')
plt.plot(y4, 'green')
plt.plot(y5, 'orange')
plt.show()
```

```python
# cas positive
y = p[1][0]
y2 = p[1][1]
y3 = p[1][2]
y4 = p[1][3]
y5 = p[1][4]
# Plot lists and show them
plt.figure(1, figsize=(16, 4)) 
plt.plot(y, 'red')
plt.plot(y2, 'blue')
plt.plot(y3, 'violet')
plt.plot(y4, 'green')
plt.plot(y5, 'orange')
plt.show()
```
